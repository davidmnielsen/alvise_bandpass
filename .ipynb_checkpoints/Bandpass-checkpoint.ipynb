```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

### Load data

```python
# raw data
path='/work/uo1075/u241292/outputs/alvise_bandpass/'
ds = xr.open_dataset(path+'E5pl00_1H_XXXX-XX_daily_250hPa_EKE_mergetime.nc')
myvar = ds['var131'][:,0,:,:].values
lat = ds['lat']
lon = ds['lon']
time = ds['time']
# ds.info()
```

```python
# Let's check what regions have high variance
plt.pcolormesh(lon, lat, np.std(myvar, axis=0))
plt.colorbar()
plt.scatter(-37,40, marker='o', c='r')
plt.xlabel('lon')
plt.ylabel('lat')
plt.show()
# Looks like the region around 37W and 40N, for example, varies a lot
```

```python
# Make time series of that point
pointCoords = [40, -37] # [lat, lon]
idlat = np.where(abs(pointCoords[0]-lat)==min(abs(pointCoords[0]-lat)))[0]
idlon = np.where(abs(pointCoords[1]-lon)==min(abs(pointCoords[1]-lon)))[0]
ts = np.squeeze(myvar[:, idlat, idlon])

plt.plot(ts)
plt.show()
```

### Filter with Scipy

```python
# https://scipy-cookbook.readthedocs.io/items/ButterworthBandpass.html
from scipy.signal import butter, lfilter

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y

ts_filt = butter_bandpass_filter(ts, 1/30, 1/10, 1, order=6)
plt.plot(ts_filt)
plt.show()
```

### Load data filtered with CDO

```python
# CDO-filtered data
#### this file is not continuous in time! CDO complained!!
ds = xr.open_dataset(path+'E5pl00_1H_XXXX-XX_daily_250hPa_EKE_mergetime_filt.nc')
myvar_CDOfilt = ds['var131'][:,0,:,:].values
ts_CDOfilt = np.squeeze(myvar_CDOfilt[:, idlat, idlon])
plt.plot(ts_CDOfilt)
plt.show()
```

### Compare

#### Frequencies

```python
import pystuff as ps
freq,   psd,   pctl,   max5,   meanRed   = ps.periods(ps.standardize(ts), dt=1, nsim=1000)
freq_f, psd_f, pctl_f, max5_f, meanRed_f = ps.periods(ps.standardize(ts_filt), dt=1, nsim=1000)
freq_CDOf, psd_CDOf, pctl_CDOf, max5_CDOf, meanRed_CDOf = ps.periods(ps.standardize(ts_CDOfilt), dt=1, nsim=1000)
```

```python
fig = plt.figure(figsize = (12,4))

ax = fig.add_subplot(131)
plt.plot(1/freq, psd,'b', label='Spectrum')
plt.plot(1/freq,ps.runmean(pctl[:,2], window=31, fillaround=True), '--g', label='95th Pctl')
plt.plot(1/freq,ps.runmean(pctl[:,3], window=31, fillaround=True), '-g',  label='99th Pctl')
plt.xlabel('Period [days]')
plt.ylabel('Power [var$^{2}$ day]')
plt.xscale('log')
plt.title('Raw data')
for i in range(5):
    t=plt.text(1/max5[i,0],max5[i,1],'%.0f days' %round(max5[i,2]), color='blue')
    t.set_bbox(dict(facecolor='white', alpha=0.5, edgecolor='w'))
    
ax = fig.add_subplot(132)
plt.plot(1/freq_f, psd_f,'b', label='Spectrum')
plt.plot(1/freq_f,ps.runmean(pctl_f[:,2], window=31, fillaround=True), '--g', label='95th Pctl')
plt.plot(1/freq_f,ps.runmean(pctl_f[:,3], window=31, fillaround=True), '-g',  label='99th Pctl')
plt.xlabel('Period [days]')
plt.ylabel('Power [var$^{2}$ day]')
plt.title('Filtered with Python')
for i in range(3):
    t=plt.text(1/max5_f[i,0],max5_f[i,1],'%.0f days' %round(max5_f[i,2]), color='blue')
    t.set_bbox(dict(facecolor='white', alpha=0.5, edgecolor='w'))
plt.xscale('log')

ax = fig.add_subplot(133)
plt.plot(1/freq_CDOf, psd_CDOf,'b', label='Spectrum')
plt.plot(1/freq_CDOf,ps.runmean(pctl_CDOf[:,2], window=31, fillaround=True), '--g', label='95th Pctl')
plt.plot(1/freq_CDOf,ps.runmean(pctl_CDOf[:,3], window=31, fillaround=True), '-g',  label='99th Pctl')
plt.xlabel('Period [days]')
plt.ylabel('Power [var$^{2}$ day]')
plt.title('Filtered with CDO')
for i in range(3):
    t=plt.text(1/max5_CDOf[i,0],max5_CDOf[i,1],'%.0f days' %round(max5_CDOf[i,2]), color='blue')
    t.set_bbox(dict(facecolor='white', alpha=0.5, edgecolor='w'))
plt.xscale('log')

plt.tight_layout()
plt.show()

fig.savefig('periodograms.png', dpi=300)
```
